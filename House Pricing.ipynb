```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#reading the data and printing the data
data = pd.read_csv(r"C:\Users\MADHAVAN\Python-madhavan\matplotlib\pratice\Housing.csv")
print(data)
```

```python
#cleaning of data
#removing the rows containing none value
data.dropna()
#removing the column containing none value
data.dropna(axis=1)
```

```python
#finding the sum of stories
lt=[]
count1=0
count2=0
count3=0
count4=0
for i in data["stories"]:
    if i == 1:
        count1 +=1
    if i ==2:
        count2+=1
    if i ==3:
        count3 += 1
    if i ==4:
        count4 += 1
lt=[count1,count2,count3,count4]
print(lt)
```

```python
#plotting the number of 1,2,3,4 stories
data.sort_values(by=['stories'])
h=data["stories"].unique()
h.sort()
print(h)
plt.pie(lt,labels=h,explode=[0.1,0.1,0.1,0.1])
plt.title("plotting number of stories", bbox={'facecolor':'0.8', 'pad':5})
```

```python
#finding the number of houses in mainroad
lt=[]
count1=0
count2=0
for i in data["mainroad"]:
    if i == "no":
        count1 +=1
    if i =="yes":
        count2+=1
lt=[count1,count2]
print(lt)
data.sort_values(by=['mainroad'])
h=data["mainroad"].unique()
h.sort()
#print(h)
plt.pie(lt,labels=h,explode=[0.1,0.1])
plt.title("Number of houses in mainroad", bbox={'facecolor':'0.8', 'pad':5})
```

```python
#plotting the frequency distribution of area
plt.hist(x=data["area"])
plt.title("Distribution of area")
```

```python
#plotting the area through line plot
plt.plot(data["area"])
plt.title("Distribution of area using lineplot")
```

```python
#plotting the scatter plot between area and price
sns.scatterplot(x="area",y="price",hue="mainroad",data=data)
plt.title("area vs price")
```

```python
#relation between different variables
x=data.corr()
print(x)
sns.heatmap(x,annot=True)
plt.title("Correlation between numerical variable")
```

```python
#plotting the regression line between area and price
sns.regplot(data=data,x=data["area"],y=data["price"])
plt.title("Area vs Price")
```

```python
#plotting multiple regression line between area and price using funishingstatus
sns.lmplot(data=data,x="area",y="price",hue="furnishingstatus")
plt.title("Area vs Price based on furnishingstatus")
```

```python
#plotting using barplot and seperating it by number of bedrooms
fig, ax = plt.subplots(figsize=(4, 5))
sns.barplot(data=data,x=data["furnishingstatus"],y=data["area"],hue=data["bedrooms"])
plt.title("Furnshingstatus vs Area based on bedrooms")
```

```python
#plotting barplot using airconditioning
sns.barplot(data=data,x=data["furnishingstatus"],y=data["area"],hue=data["airconditioning"])
plt.title("Furnshingstatus vs Area based on airconditioning")
```

```python
#plotting using barplot and seperating it by number of bathrooms
sns.barplot(data=data,x=data["furnishingstatus"],y=data["area"],hue=data["bathrooms"])
plt.title("Furnshingstatus vs Area based on bathrooms")
```

```python
#multiple subplots or grid in same figure
sea = sns.FacetGrid(data, col = "furnishingstatus")
sea.map(sns.barplot,x=data["furnishingstatus"],y=data["area"])
```

```python
#plotting the lineplot across area and price
sns.lineplot(data=data,x="area",y="price",hue="furnishingstatus",marker='>')
plt.title("Price vs Area using furnishingstatus")
```

```python
#A Box plot is a way to visualize the distribution of the data by using a box and some vertical lines
#It is known as the whisker plot. 
#The data can be distributed between five key ranges, which are as follows:
#Minimum,1st quartile,Median,3rd quartile,Maximum
plt.boxplot(data["area"],notch='True')
plt.title("Box plot")
```

```python
#removing the outliners from the data
precen25=data["area"].quantile(0.25)
precen75=data["area"].quantile(0.75)
iqr=precen75-precen25
lower_limit=precen25+1.5*iqr
upper_limit=precen75+1.5-iqr
data[data['area'] > upper_limit]
data[data['area'] < lower_limit]
data = data[data['area'] < upper_limit ]
plt.boxplot(data["area"],notch='True')
plt.title("Boxplot with Removed Outliners")
```
